```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("train_data_2.csv")
df.head()
```

### Check The Clasees Frequency 

```python
## Checking the frequency for each class (Price range)
sns.histplot(data=df, x="price_range", bins=7, color='royalblue', alpha=0.7)
plt.title("Price Range Distribution")
plt.xlabel("Price Range")
plt.ylabel("Frequency")
plt.show()
```

### Realtionship between battery_power and price_range

```python
plt.figure(figsize=(12, 6))
sns.swarmplot(y=df["battery_power"], x=df["price_range"])  
plt.title(f"Relationship between battery power and Price Range") 
plt.xticks([0,1,2,3])
plt.show()
```
